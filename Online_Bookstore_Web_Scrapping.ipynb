```python
import requests
import bs4
```

```python
# create an empty list for all the entities
prices=[]
pages=[]
ratings=[]
titles=[]
image_url=[]
```

```python
# Specify the number of pages you want.
no_of_pages=1
```

```python
# looping through the required pages and selecting the pages accordingly
for i in range(1,no_of_pages+1):  # +1 because we want to include the last page
 url=('https://books.toscrape.com/catalogue/page-{}.html'.format(i))
 pages.append(url)
```

```python
print(len(pages))
print(pages)
```

**Getting the data from the particular page url and converting from object of type 'request to object of type beautiful soup, making it more readable**

```python
for item in pages:
  pages=requests.get(item)
  soup=bs4.BeautifulSoup(pages.text,'html.parser')
```

```python
#print(soup.prettify())
```

**Find all the titles**

```python
# print all titles
for t in soup.find_all('h3'):
  print(t)
```

```python
# finding all the titles and appending them to empty title list.
for t in soup.find_all('h3'):
  titless=t.getText()
  titles.append(titless)
```

```python
# showing what is inside the title variable
titles
```

**Find all the prices**

```python
for p in soup.find_all('p',class_='price_color'):
  price=p.getText()
  prices.append(price)
```

```python
# Shwoing whats inside prices
prices
```

**Finding all the ratings**

```python
# getting the ratings
for s in soup.find_all('p',class_='star-rating'):
  for k,v in s.attrs.items():  #read about attrs
    star=v[1]
    ratings.append(star)
    #print(star) #
```

```python
# showing what is in the ratings
ratings
```

**Find all the image URL's**

```python
#fetching all the div tags in the class called image_container
#print(divs)
divs=soup.find_all('div',class_='image_container')
for thumbs in divs:
  tagss=thumbs.find('img',class_='thumbnail')
  #print(tagss)
  links='https://books.toscrape.com/' +str(tagss['src'])
  new_links=links.replace('..','') #to get rid of the three-dost, u can check it in your browser to see if we ignore it, the link still works, besides ignoring it makes the link more readable
  image_url.append(new_links) #our url list now contains all of the urls of the book images
```

```python
# showing what is inside the url
image_url
```

**Processing the scrapped data**

```python
# convert the list into a dataFrame
# create adictionary that will cntain the data from the site
web_data={'Title':titles,
         'Price':prices,
         'Ratings':ratings,
         'image_URL':image_url}
```

```python
import pandas as pd
#make sure all your arrays are of same lenght, Note that, each time you run the
# cells above, it will continue fetaching the data from the webpage and chances
#are you have refetched the data couple of time. Just rerun all the above cells
#once starting from where you created the empty list for title,urls,prices and ratings.
print(len(titles))
print(len(prices))
print(len(ratings))
print(len(image_url))
```

```python
df=pd.DataFrame(web_data)
#df
```

```python
#  changing the index to tart from 1. not really necessary

df.index+=1 #adds 1 to all the current index
#df
```

```python
# getting the currency sign attached with the price away.
df['Price']=df['Price'].str.replace('Â£','')
df
```

```python
# sorting the price column to see the book with the highest price
df.sort_values(by='Price', ascending=False,inplace=True)
```

**Converting the rating colimn to an integer instead of a string**

```python
df['Ratings']=df['Ratings'].replace({'One':1, 'Two':2, 'Three':3, 'Four':4, 'Five':5})
df
```

```python
# checking the data types of all variables
df.dtypes
```

```python
# convert price to float
df['Price']=df['Price'].astype(float)
```

**is there any correlation between rating and price?**

```python
corr_p_r=df[['Ratings','Price']].corr()
corr_p_r
```

```python
# visualizing the correlation with a heatmap
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(5,5))
sns.heatmap(corr_p_r,annot=True)
```
